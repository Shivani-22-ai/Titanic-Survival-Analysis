# Titanic Survival Prediction
-----------------------------------
### Objective
Build a machine learning model to predict whether a passenger survived the Titanic disaster.

-------------------------------------------------------------------------------------------------------------
### Importing libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

-------------------------------------------------------------------------------------------------------------
### Loading dataset

```python
titan = pd.read_csv(r'C:\Users\chand\Downloads\Titanic-Dataset.csv')
titan.head()
```

----------------------------------
### Data Preprocessing
- Missing value analysis
- Feature Engineering
- Encoding


```python
# Missing values analysis
titan.isnull().sum()
```

```python
# Filling the missing values
titan['Age'] = titan.Age.fillna(value = titan.Age.median())
titan['Embarked'] = titan.Embarked.fillna(value = titan.Embarked.mode()[0])
titan = titan.drop('Cabin',axis=1)
titan.isnull().sum()
```

```python
# Feature Engineering
titan["Family_size"] = titan["SibSp"] + titan["Parch"]
titan.groupby("Family_size")["Survived"].sum()
```

```python
# Encoding
titan["Sex"] = titan["Sex"].map({"male":0,"female":1})
titan
```

-----------------------------------------------
### Feature Selection and train-test split
Features:
- Pclass
- Age
- Fare
- Sex
- Family_Size

Target:
- Survived


```python
X = titan[["Pclass","Age","Fare","Sex","Family_size"]]
y = titan["Survived"]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

---------------------------------------------------------------------------
### Model training and evaluation

```python
dm = DecisionTreeClassifier(random_state=5)
dm.fit(X_train,y_train)
pre = dm.predict(X_test)
acc = accuracy_score(pre,y_test)
print("Decision Tree Accuracy:",acc)
```

-------------------------------------
### Feature Importance

```python
dm.feature_importances_
```

----------------------------
## Conclusion (Version 1)
The Decision Tree model achieved an accuracy of 80.45%. Among the selected features, Sex and Pclass were the most influential predictors of survival. The project demonstrates a complete machine learning workflow including preprocessing, feature engineering, model training, and evaluation.
